# **Import SparkSession**

```python
from pyspark.sql import SparkSession
```

# **Create Spark Session**

```python
spark = SparkSession.builder \
    .appName("Employees Analysis") \
    .getOrCreate()
```

# **Load Employees CSV File**

```python
df = spark.read.csv("employees.csv", header=True, inferSchema=True)
```

# **View Data**

```python
df.show()
```

# **Check Schema**

```python
df.printSchema()
```

# **Count Rows**

```python
df.count()
```

# **Select Specific Columns**

```python
df.select("First Name", "Salary").show()
```

# **Renaming column**

```python
#Renaming First Name
df = df.withColumnRenamed("First Name", "First_Name")
```

```python
# Renaming Start Date
df = df.withColumnRenamed("Start Date", "Start_Date")
```

```python
# Renaming Last Login Time
df = df.withColumnRenamed("Last Login Time", "Last_Login_Time")
```

```python
# Renaming Bonus %
df = df.withColumnRenamed("Bonus %", "Bonus")
```

```python
# Renaming Senior Management
df = df.withColumnRenamed("Senior Management", "Senior_Management")
```

```python
# view data
df.show()
```

# **Filter Data**

```python
# Employees with salary above 50000
df.filter(df.Salary > 50000).show()
```

# Multiple Conditions

```python
df.filter((df.Team == "Finance") & (df.Salary > 40000)).show()
```

# **sort Data**

```python
#Highest Salary First
df.orderBy(df.Salary.desc()).show()
```

# **Group By Department**

```python
# Average Salary by Department
from pyspark.sql.functions import avg

df.groupBy("Team").agg(avg("Salary")).show()
```

# **Count Employees per Team**

```python
df.groupBy("Team").count().show()
```

# **Handle Missing Values**

```python
# Check Nulls
df.show()
```

```python
# Fill Null Salary with 0
df.fillna({"Salary":0}).show()
```

```python
df.printSchema()
```

```python
from pyspark.sql.functions import col, when, count

df.select([
    count(when(col(c).isNull(), c)).alias(c)
    for c in df.columns
]).show()
```

```python
# check missing values

from pyspark.sql.functions import trim

df.select([
    count(when(trim(col(c)) == "", c)).alias(c)
    for c in df.columns
]).show()
```

# **Handle String Columns**

```python
#Fill nulls with meaningful text:
df = df.fillna({
    "First_Name": "Unknown",
    "Team": "Unknown"
  })
```

```python
df = df.fillna({
    "Gender": "Unknown",
    "Last_Login_Time": "Unknown"

  })
```

```python
#Handle Numeric Columns
#Fill With Median
median_salary = df.approxQuantile("Salary", [0.5], 0)[0]
median_bonus = df.approxQuantile("Bonus",[0.5],0)[0]
df = df.fillna({
    "Salary": median_salary,
    "Bonus": median_bonus
})
```

```python
# Handle Date Columns
# If Start_Date has nulls:

df = df.fillna({
    "Start_Date": "Unknown"
})
```

```python
df = df.replace("######", None)
df = df.replace("N/A", None)
```

# handling missing value in Senior_Management

```python
df.groupBy("Senior_Management").count().show()


# fill with Null with mode
mode_value = (
    df.groupBy("Senior_Management")
      .count()
      .orderBy("count", ascending=False)
      .first()[0]
)

print(mode_value)

#fill null
df = df.fillna({"Senior_Management": mode_value})
```

```python
# recheck missing value
df.select([
    count(when(col(c).isNull(), c)).alias(c)
    for c in df.columns
]).show()
```

# **SQL Queries in PySpark**

```python
#Create temp table:
df.createOrReplaceTempView("employees")
```
